## 2. Reading data into Stata
We can import data into Stata from the current directory.  
* use - .dta file. (data in Stata format)
* insheet - ASCII text file. (csv, txt)

The option **clear** can delete the current dataset in Stata.

```stata
use "file_name.dta", clear // import .dta file
```

```stata
insheet using "affairs.csv", clear // import ASCII text file
```

* save "saveing directory/file_name.dta", replace

## 3. Examining the data
* describe - # of observations and variables, and types of variables.
* summarize - summary statistics of variables. (means, standard deviations)
* list - raw data.
* tab - details of variables.
* inspect - distribution of variables.

```stata
describe
```

```stata
summarize // or sum
```

```stata
list
```

## 4. Keeping track of things
* do-file - code.
* log-file - resutls.
* label - add description to data and variables.
* label define & label value - add description to the value of variables.

```stata
capture log close // close the log file if any
log using affairs_ta0226.log, replace text 
// log using - open a log file 
// replace - overwrites any log file of the same name
// text - overrides the default and makes the log file readable in any editor
```

```stata
describe // original description of data
```

```stata
label data "affairs data" // label the description of data
```

```stata
// label the description of each variable
label variable id "identifier"
lab var male "=1 if male"
lab var age "in years"
lab var yrsmarr "years married"
lab var kids "=1 if have kids"
lab var relig "5 = very relig, 4 = somewhat, 3 = slightly, 2 = not at all, 1 = anti"
lab var educ "years schooling"
lab var occup "occupation, reverse Hollingshead scale"
lab var ratemarr "5 = vry hap marr, 4 = hap than avg, 3 = avg, 2 = smewht uphap, 1 = vry unhap"
lab var naffairs "number of affairs within last year"
lab var affair "=1 if had at least one affair"
```

```stata
describe // the description of data after labeling
```

```stata
tab male // original tab of var male
```

```stata
label define sex 0 "female" 1 "male" // define sex as 0 with label "female", 1 with lable "male"
label values male sex // give var male the definition of sex
```

```stata
tab male
```

```stata
describe
```

```stata
label list sex // check the value of label "sex"
```

```stata
log close
```
